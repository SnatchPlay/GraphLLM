````python
import os
from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv())
import time
from langchain_groq import ChatGroq
import streamlit as st
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain, GraphCypherQAChain, RetrievalQA
from langchain_community.vectorstores.neo4j_vector import Neo4jVector
from langchain.chains.conversation.memory import ConversationBufferMemory
from langchain.agents import AgentExecutor, create_react_agent
from langchain.tools import Tool
from langchain_community.graphs import Neo4jGraph
from langchain_community.embeddings import HuggingFaceEmbeddings
from IPython.display import Markdown
from langchain_community.embeddings import GPT4AllEmbeddings
from langchain_openai import ChatOpenAI
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_community.llms import NLPCloud
from langchain_community.embeddings import NLPCloudEmbeddings
import neo4j
import sentence_transformers
````

````python
API_KEY = "there is no api key"#groq
#API_KEY=" "#openai
NLP_API_KEY=" there is no api key"#nlp

#Google_API_KEY=" "#google
#NEO4J_URI='bolt://0.tcp.eu.ngrok.io:19792'
NEO4J_URI='bolt://localhost:7687'
NEO4J_USERNAME='neo4j'
NEO4J_PASSWORD='11111111'
````

````python
llm = ChatGroq(temperature=0, groq_api_key=API_KEY, model_name="mixtral-8x7b-32768")
#llm = ChatOpenAI(model="gpt-3.5-turbo", temperature=0, api_key=API_KEY)
#llm = NLPCloud(nlpcloud_api_key=API_KEY, model_name="finetuned-llama-3-70b")
````

````python
graph = Neo4jGraph(url=NEO4J_URI,username=NEO4J_USERNAME,password=NEO4J_PASSWORD)
````

````python
cypher_chain = GraphCypherQAChain.from_llm(llm,graph=graph,verbose=True)
````

````python
#embeddings = HuggingFaceEmbeddings()
#embeddings=GoogleGenerativeAIEmbeddings(google_api_key=Google_API_KEY,model="models/embedding-001")
os.environ["NLPCLOUD_API_KEY"] = NLP_API_KEY
embeddings = NLPCloudEmbeddings() 
````

````python
neo4jvector = Neo4jVector.from_existing_relationship_index(
    embeddings,
    url=NEO4J_URI,
    username=NEO4J_USERNAME,
    password=NEO4J_PASSWORD,
    index_name="relation_embedding"
)
````

````python
retriever = neo4jvector.as_retriever()

kg_qa = RetrievalQA.from_chain_type(
    llm,                  # <1>
    chain_type="stuff",   # <2>
    retriever=retriever,  # <3>
)
````

````python
tools = [
    Tool.from_function(
        name="General Chat",
        description="For general chat not covered by other tools",
        func=llm.invoke,
        return_direct=True,
        handle_parsing_errors=True
    ),
    Tool.from_function(
        name="Vector Search Index",
        description="Provides information using Vector Search",
        func = kg_qa,
        return_direct=False,
        handle_parsing_errors=True
    ),
    Tool.from_function(
        name="Cypher QA",
        description="Provide information about questions using Cypher",
        func = cypher_chain,
        return_direct=False,
        handle_parsing_errors=True
    )
]
````

````python
# tag::prompt[]
CYPHER_GENERATION_TEMPLATE = """
You are an expert Neo4j developer providing information from the ICIJ Offshore Leaks Database. You translate user's requests into Cypher to answer questions about offshore entities, officers, intermediaries, and addresses. DO NOT use any other relationship types or properties that are not provided.

DO NOT answer any questions using your pre-trained knowledge! ONLY use the information provided in the context!

TOOLS:
You have access to the following tools: {tools}

To use a tool, please USE the following format:
```
Thought: Do I need to use a tool? Yes
Action: the action to take, should be one of [{tool_names}]
Action Input: the input to the action
Observation: the result of the action
```
Example Cypher Statements:

1. Could you provide a list of all entities registered in the British Virgin Islands?
MATCH (e:Entity)-[:registered_address]->(a:Address)
WHERE a.country_codes = "VGB"
RETURN e.name

2. Who are the officers connected to this company?
MATCH (e:Entity)-[:connected_to]->(o:Officer) 
WHERE e.name = 'Company Name' 
RETURN o.name

3. What other entities is this intermediary associated with?
MATCH (i:Intermediary)-[:connected_to]->(e:Entity) 
WHERE i.name = 'Intermediary Name' 
RETURN e.name

When you have a response to say to the Human, or if you do not need to use a tool, you MUST use the format:
```
Thought: Do I need to use a tool? No
Final Answer: [your response here]
```
Note: DO NOT include any explanations or apologies in your responses.
If the tool did not provide a suitable answer or the question IS NOT related to your speciality, you should say "I don't know" or "I can't answer the question".
```
Final Answer: I don't know
```
Begin!
Previous conversation history:
{chat_history}

New input: {input}
{agent_scratchpad}
"""
# end::prompt[]
````

````python
# tag::template[]
agent_prompt = PromptTemplate.from_template(CYPHER_GENERATION_TEMPLATE)
# end::template[]
agent = create_react_agent(llm, tools, agent_prompt)
memory = ConversationBufferMemory(memory_key='chat_history',k=5,return_messages=False)
````

````python
agent_executor = AgentExecutor(
    agent=agent,
    tools=tools,
    memory=memory,
    verbose=True,
    handle_parsing_errors=True
    )
````

````python
agent_executor.invoke({"input": "Generate a concise report on the geographical distribution of entities and activities revealed in the Offshore Leaks."})['output']
````

````python
agent_executor.invoke({"input": "Could you provide a list of all entities registered in the British Virgin Islands?"})['output']
````

````python
agent_executor.invoke({"input": "What are the most common countries of origin for officers connected to offshore entities?"})['output']
````
